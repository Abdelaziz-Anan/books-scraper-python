```python
URL='http://books.toscrape.com/catalogue/category/books/mystery_3/index.html'
```

```python
def get_details_of_books(page) :
    with open ('Books.csv','w', newline='') as file :
        writer = csv.DictWriter(file , fieldnames=['Title','Rating','Price'])
        writer.writeheader()
        soup = BeautifulSoup(requests.get(page).content ,'html.parser')
        all_article = soup.find_all('article',attrs = {'class':'product_pod'})
        rate = {'One':1 ,'Two':2 ,'Three':3 ,'Four':4 ,'Five' : 5 ,'Six' : 6 , 'Seven' : 7 , 'Eight' : 8}
        for article in all_article:
            
            book_name= article.find('h3').find('a').text
            rating = article.find('p', attrs ={'class':"star-rating"}).attrs['class'][1]
            price = float(article.find('div',attrs ={'class' : 'product_price'}).find('p',attrs={'class': 'price_color'}).get_text().split('£')[1])

            writer.writerow({'Title' : book_name ,'Rating' : rate[rating] ,'Price' : price })
            
            print(f'Book name:\n{book_name}\n\nPrice:\n{price}\n\nRating:\n{rate[rating]}\n\n{'-'*70}')
```

```python
get_details_of_books(URL)
```
